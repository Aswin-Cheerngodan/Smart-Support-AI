```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv(r"D:\Capstone Projects\Smart-Support-AI\data\knowledge base data\Customer_Support_Conversations.csv")

df.head()
```

```python
def query_extract(text):
    match = re.search(r'Customer: (.*?)\n', text, re.DOTALL)
    first_query = match.group(1) if match else None
    return first_query
```

```python
df['query'] = df['conversation'].apply(query_extract)
```

```python
df_clean = df.dropna()
```

```python
df_clean = df[['issue_area', 'query', 'conversation']]

df_clean.head()
```

```python
df_clean.to_csv(r"D:\Capstone Projects\Smart-Support-AI\data\knowledge base data\knowledge_base.csv", index=False)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import time
model = SentenceTransformer('all-MiniLM-L6-v2')  # Fast and efficient model

text1 = "In her leisure time, she likes to read novels."
text2 = "She enjoys reading books in her free time."
print("start")
start = time.time()
embeddings = model.encode([text1, text2])
similarity = np.dot(embeddings[0], embeddings[1]) / (np.linalg.norm(embeddings[0]) * np.linalg.norm(embeddings[1]))
end = time.time()
print("end")
print(f"Similarity Score: {similarity:.4f}  time {end-start}")
```
